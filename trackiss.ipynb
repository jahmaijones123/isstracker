```python
import json 
import turtle 
import urllib.request 
import time 

url = 'http://api.open-notify.org/astros.json'
response = urllib.request.urlopen(url)

result = json.loads(response.read())
print(result)

print('People in space:', result['number'])

people = result ['people']
print(people)

for x in people:
    print(x)

for x in people:
    print(x ['name'], 'in', x ['craft'])


# shows current location of the ISS
url = 'http://api.open-notify.org/iss-now.json' 

response = urllib.request.urlopen(url)
result = json.loads(response.read())
print(result)

location = result ['iss_position']
lat = location ['latitude']
lon = location ['longitude']

print ('latitude', lat)
print ('location', lon)

lat_new = float(lat)
lon_new = float(lon)

screen = turtle.Screen()

screen.setup(720, 360)
screen.setworldcoordinates(-180,-90,180,90)
screen.register_shape('C:\\Users\\Jahmai\\Pictures\\iss\\iss.gif')
screen.bgpic('C:\\Users\\Jahmai\\Pictures\\iss\\map.gif')

iss = turtle.Turtle()
iss.shape('C:\\Users\\Jahmai\\Pictures\\iss\\iss.gif')
iss.setheading(90)
iss.penup()

iss.goto(lat_new, lon_new)

lat = float (52.4117)
lon = float (-1.9618)

location = turtle.Turtle()
location.penup()
location.color('yellow')
location.goto(lon,lat)
location.dot(5)
location.hideturtle()

url = 'http://api.open-notify.org/iss-pass.json?lat=' +str(lat) +'&lon=' +str(lon)
response = urllib.request.urlopen(url)
result = json.loads(response.read())
turtle.done()

# print result 

over = result['response'][1]['risetime']
location.write(time.ctime(over))



   




```
